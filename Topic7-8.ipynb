#### 5. Pie Chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()

df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize =(10,8),
                            startangle=90,
                            autopct='%1.1f%%'
                           )

plt.title('Immigration to Canada by Continent')
plt.axis('equal')

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral','lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                           figsize =(15,6),
                            startangle=90,
                            autopct='%1.1f%%',
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list
                           )

plt.title('Immigration to Canada by Continent', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

#### 7. Scatter Plot

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

df_tot.index = map(int, df_tot.index)

df_tot.reset_index(inplace=True)

df_tot.columns = ['year', 'total']

df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize = (10,7), color='green')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x =df_tot['year']
y =df_tot['total']

fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize = (10,7), color='green')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.plot(x, fit[0] * x + fit[1], color='blue')

plt.show()

'No. Immigrants = {0:.0f} * year + {1:.0f}'.format(fit[0], fit[1])
```

```python
df_incidents = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
```

```python
df_incidents.head()
```

```python
df_incidents.shape
```

```python
limit = 100

df_incidents = df_incidents.iloc[0:limit, :]
```

```python
df_incidents.shape
```

```python
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(locations=[latitude, longitude], zoom_start=12)
```

#### Maps with Markers

```python
incidents = folium.map.FeatureGroup()

for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
sanfran_map.add_child(incidents) 
```

```python
incidents = folium.map.FeatureGroup()

for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            popup=label,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
sanfran_map.add_child(incidents)
```
